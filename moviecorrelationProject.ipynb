```python
# First let's import the packages we will use in this project
# You can do this all now or as you need them
import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

pd.options.mode.chained_assignment = None

```

```python
# Read in the dataset
df = pd.read_csv(r"movies.csv")
```

```python
#Check for missing data
for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, round(pct_missing*100)))
```

```python
#Change data types of columns
df['budget'] = pd.to_numeric(df['budget'], errors='coerce') 
df['budget'] = df['budget'].fillna(0).astype('int64')
df['gross'] = pd.to_numeric(df['gross'], errors='coerce') 
df['gross'] = df['gross'].fillna(0).astype('int64')

```

```python
# Creating the correct year column
df['yearcorrect'] = df['released'].astype(str).str[:4]
```

```python
#Check for outliers
df.boxplot(column=['gross'])
```

```python
#Drop duplicates
df.drop_duplicates()
```

```python
#Sort the data 
df.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
#Regression plot
sns.regplot(x="gross", y="budget", data=df)
```

```python
sns.regplot(x="score", y="gross", data=df)
```

```python
#Pearson Correlation plot between all numeric values
df_numeric = df.select_dtypes(include=['float64', 'int64'])
df_numeric.corr(method='pearson')
```

```python
#kendal Correlation plot between all numeric values
df_numeric = df.select_dtypes(include=['float64', 'int64'])
df_numeric.corr(method='kendall')
```

```python
#Pearson Correlation plot between all numeric values
df_numeric = df.select_dtypes(include=['float64', 'int64'])
df_numeric.corr(method='spearman')
```

```python
df_numeric = df.select_dtypes(include=['float64', 'int64'])
correlation_matrix = df_numeric.corr()

sns.heatmap(correlation_matrix, annot = True)

plt.title("Correlation matrix for Numeric Features")

plt.xlabel("Movie features")

plt.ylabel("Movie features")

plt.show()
```

```python
# Using factorize assign a random numeric value for each unique categorical value
df.apply(lambda x: x.factorize()[0]).corr(method='pearson')
```

```python
correlation_matrix = df.apply(lambda x: x.factorize()[0]).corr(method='pearson')

sns.heatmap(correlation_matrix, annot = True)

plt.title("Correlation matrix for Movies")

plt.xlabel("Movie features")

plt.ylabel("Movie features")

plt.show()
```

```python
correlation_mat = df.apply(lambda x: x.factorize()[0]).corr()

corr_pairs = correlation_mat.unstack()

print(corr_pairs)
```

```python
sorted_pairs = corr_pairs.sort_values(kind="quicksort")

print(sorted_pairs)
```

```python
#Checks those with high correlation (> 0.5)

strong_pairs = sorted_pairs[abs(sorted_pairs) > 0.5]

print(strong_pairs)
```
